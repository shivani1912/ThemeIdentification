```python
#getting the data from files
import string
import re
feedback_df=pd.DataFrame(columns=["no_of_votes","thread_title","thread_link","description"])
a=list(range(1,100))
for number in a:
        filename = 'feedback_pg'+str(number)+'.htm'
        with open(filename, 'r', encoding='utf-8') as f:
            text = f.read()
            soup = BeautifulSoup(text, 'html.parser') 
            all_feedback=soup.find_all('li', attrs={'class':'uvListItem uvIdea uvIdea-list'})
            for feedback in all_feedback:
                #no_of_votes
                no_of_votes=feedback.find('div', attrs={'class':'uvIdeaVoteCount'}).get_text()
                no_of_votes=re.sub('[A-Z|a-z|]+', '',no_of_votes)
                no_of_votes=re.sub('\\ *', '',no_of_votes)
                temp = re.findall(r'\d+', no_of_votes) 
                no_of_votes= list(map(int, temp)) 
                #print(no_of_votes)
                #thread_title
                thread_title=feedback.find('h2', attrs={'class':'uvIdeaTitle'}).get_text()
                thread_title=thread_title.strip()
                thread_title=thread_title.replace('\n', ' ')
                thread_title=re.findall('[A-Z|a-z|]+',thread_title)
                
                #print(thread_title)
                #thread_link
                thread_link=feedback.find('h2', attrs={'class':'uvIdeaTitle'})
                thread_link=thread_link.find('a', href=True)
                thread_link=thread_link['href']
                
                #description
                if(feedback.find('div', attrs={'class':'typeset'}) is None):
                    description=''
                else:
                    description=feedback.find('div', attrs={'class':'typeset'}).get_text()
                    description=re.sub('\\xe2\\x80\\x99','',description)
                #print(description)
                
               
                
                new_row=pd.DataFrame([[no_of_votes,thread_title,thread_link,description]],
                                columns=["no_of_votes","thread_title","thread_link","description"])
                feedback_df=feedback_df.append(new_row)
                
                print("\n")
```

```python
negative_comments=feedback_df[feedback_df['sentiment']<0.5].sort_values(by=['no_of_votes'],ascending=False).head(300)
```

```python
#NLTK Stop words
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
import re

# Convert to list
data = negative_comments.allcomments.values.tolist()


# Remove Emails
data = [re.sub('\S*@\S*\s?', '', sent) for sent in data]

# Remove new line characters
data = [re.sub('\s+', ' ', sent) for sent in data]

# Remove distracting single quotes
data = [re.sub("\'", "", sent) for sent in data]

```

```python
import re
import numpy as np
import pandas as pd
from pprint import pprint

# Gensim
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel

# spacy for lemmatization
import spacy

# Plotting tools
import pyLDAvis
import pyLDAvis.gensim  # don't skip this
import matplotlib.pyplot as plt
%matplotlib inline

# Enable logging for gensim - optional
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.ERROR)

import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)
```

```python
#tokenizing & removing punctuations
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations

data_words = list(sent_to_words(data))
```

```python
#remove stopwords
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

```python
data_words_nostops = remove_stopwords(data_words)
data_lemmatized = lemmatization(data_words_nostops, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])

print(data_lemmatized[1:5])
```

```python
# Create Dictionary
id2word = corpora.Dictionary(data_lemmatized)

# Create Corpus
texts = data_lemmatized

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# View
print(corpus[:1])
```

```python
# Build LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=10, 
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,
                                           passes=10,
                                           alpha='auto',
                                           per_word_topics=True)
```

```python
# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```

```python
# Visualize the topics
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, corpus, id2word)
vis
```
